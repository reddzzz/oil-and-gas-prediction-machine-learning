```python
import numpy as np
import pandas as pd
from scipy import stats
from sklearn.metrics import mean_squared_error
from math import *
import matplotlib.pyplot as plt
from hyperopt import tpe, hp, fmin, STATUS_OK,Trials
from hyperopt.pyll.base import scope
from sklearn.ensemble import *
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler 
```

```python
data = pd.read_excel("/Users/sreeya/Documents/Uni/Thesis/new kyle n volve/final_dataset2.xlsx")
data1 = pd.read_excel("/Users/sreeya/Documents/Uni/Thesis/new kyle n volve/final_dataset2.xlsx")
```

Oil

```python
X = data1.filter(['Hours Online', 'Av. DHT (Deg C)', 'Av. DHP (bar)','Av. WHP (bar)', 'Av. WHT (Deg C)'])
Y = data['Oil (m3)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
```

```python
space = {
    "n_estimators": scope.int(hp.quniform('n_estimators', 5, 100, 1)),
    "min_samples_split": scope.int(hp.quniform('min_samples_split', 2, 50, 1)),
    "max_depth": scope.int(hp.quniform('max_depth', 2, 50, 1)),
}
```

```python
def hyperparameter_tuning(space):
    model=RandomForestRegressor(n_estimators =space['n_estimators'], max_depth = int(space['max_depth']), 
                                    min_samples_split = space['min_samples_split'])
    model.fit(X_train, y_train)

    pred = model.predict(X_val)
    rmse = sqrt(mean_squared_error(y_val, pred))
    print ("SCORE:", rmse)
    return {'loss':rmse, 'status': STATUS_OK, 'model': model}
```

```python
trials = Trials()
best = fmin(fn=hyperparameter_tuning,
            space=space,
            algo=tpe.suggest,
            max_evals=1000,
            trials=trials)

print (best)
```

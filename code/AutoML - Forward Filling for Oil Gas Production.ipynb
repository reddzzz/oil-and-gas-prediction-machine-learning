```python
!pip3 install openpyxl
!pip3 install matplotlib
!pip3 install h2o
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
data1 = pd.read_excel("/Users/sreeya/Documents/Uni/Thesis/dataset_kyle.xlsx")
data2 = pd.read_excel("/Users/sreeya/Documents/Uni/Thesis/dataset2.xlsx")
data = pd.concat([data1,data2], ignore_index=True)
```

```python
data.drop(data1.columns[[0]], axis=1, inplace=True)
```

```python
data_f = h2o.H2OFrame(data)
```

```python
train, valid = data_f.split_frame(ratios = [.9])
```

```python
testing_data = pd.read_excel("/Users/sreeya/Documents/Uni/Thesis/cleaned/cleaned.xlsx")
```

```python
testing_data = testing_data.dropna()
```

```python
testing_data.isnull().sum()
```

```python
data_pred = h2o.H2OFrame(testing_data)
```

```python
x = train.columns
y = "Oil (m3)"
x.remove(y)
x.remove('Gas (m3)')
```

```python
aml = H2OAutoML(max_runtime_secs = 1800, sort_metric = "RMSE",nfolds = 0)
aml.train(x=x, y=y, training_frame=train,validation_frame=valid)
```

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```

```python
h = h2o.get_model("GBM_grid_1_AutoML_1_20220527_154746_model_715")
```

```python
preds = h.predict(data_pred)
```

```python
preds
```

```python
data_as_list = h2o.as_list(preds, use_pandas=False)
pr = h2o.as_list(preds)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

#calculate RMSE
sqrt(mean_squared_error(pr['predict'], testing_data['Oil (m3)'])) 
```
